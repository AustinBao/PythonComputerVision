```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('C:/Users/Austi/Downloads/DATA/horse.jpg') # BGR for cv2
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB) # RGB for display

rainbow = cv2.imread('C:/Users/Austi/Downloads/DATA/rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('C:/Users/Austi/Downloads/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
# index 0 is blue on cv2 (B G R); mask is if you want to only calculate a section of the img
histval = cv2.calcHist([blue_bricks], channels=[0], mask=None,histSize=[256], ranges=[0,256])
plt.plot(histval)
```

```python
histval = cv2.calcHist([dark_horse], channels=[0], mask=None,histSize=[256], ranges=[0,256])
plt.plot(histval)
```

```python
# HOW TO SHOW ALL COLOR CHANNELS ON A HISTOGRAM    *IMPORTANT

img = blue_bricks
color = ("b","g","r")

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])
    
plt.title("Blue brick histogram")
```

```python
img = dark_horse
color = ("b","g","r")

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr, color=col)
# xlim and ylim determine the scale of the graph shown. Lower the ranges to zoom in.
    plt.xlim([0,50])
    plt.ylim([0,500000])
    
plt.title("Dark horse histogram")
```

```python
rainbow = cv2.imread('C:/Users/Austi/Downloads/DATA/rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
img = rainbow
img.shape
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
plt.imshow(mask, cmap="gray")
```

```python
# mask object that we created (white triangle)
mask[300:400, 100:400] = 255
```

```python
plt.imshow(mask, cmap="gray")
```

```python
masked_img = cv2.bitwise_and(img,img,mask=mask) # BGR for histogram calc
```

```python
showmaskedimg = cv2.bitwise_and(show_rainbow, show_rainbow, mask=mask) # for visuals
```

```python
plt.imshow(showmaskedimg)
```

```python
# BGR ; with mask
histmaskval = cv2.calcHist([rainbow], channels=[2], mask = mask, histSize=[256], ranges=[0,256])
```

```python
# without mask
histvalred = cv2.calcHist([rainbow], channels=[2], mask = None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(histmaskval)
plt.title("Red histogram for masked rainbow")
```
