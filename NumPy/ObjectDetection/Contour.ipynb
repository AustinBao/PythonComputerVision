```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('C:/OpenCV/DATA/internal_external.png', 0) 
plt.imshow(img,cmap="gray")
```

```python
image, contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
externalcontours = np.zeros(image.shape)
externalcontours.shape
```

```python
for i in range(len(contours)):

    #External if -1
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(externalcontours, contours, i , 255, -1)
```

```python
plt.imshow(externalcontours, cmap="gray")
```

```python
internalcontour = np.zeros(image.shape)

for i in range(len(contours)):

    #Internal if not -1
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internalcontour, contours, i , 255, -1)

plt.imshow(internalcontour, cmap="gray")
```

```python
smilecontour = np.zeros(image.shape)

for i in range(len(contours)):

    #Can differentiate and choose which internal contour we want. In this case, 0 is the smiley face in the hierarchy list  
    if hierarchy[0][i][3] == 0:
        cv2.drawContours(smilecontour, contours, i , 255, -1)

plt.imshow(smilecontour, cmap="gray")
```

```python
# same with the pepperoni slices on the pizza

pepperonicontour = np.zeros(image.shape)

for i in range(len(contours)):

    if hierarchy[0][i][3] == 4:
        cv2.drawContours(pepperonicontour, contours, i , 255, -1)

plt.imshow(pepperonicontour, cmap="gray")
```
